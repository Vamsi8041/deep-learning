```python
import numpy as np

def step(sum):
    if sum >= 0:
        return 1
    return 0

def perceptron(x, y):
    l = 0.1
    w = np.zeros(x.shape[1])
    b = 0
    for ep in range(10):
        for i in range(x.shape[0]):
            sum = np.dot(w, x[i]) + b
            y_pred = step(sum)
            e = y[i] - y_pred
            if e != 0:
                w = w + l * e * x[i]
                b = b + l * e
            print(f"Epoch {ep+1}, Sample {i+1}: w={w}, b={b}")
    return w, b

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

w, b = perceptron(x, y)

for i in range(x.shape[0]):
    sum = np.dot(w, x[i]) + b
    y_pred = step(sum)
    print(f"Input: {x[i]}, Predicted: {y_pred}, Actual: {y[i]}")
```
